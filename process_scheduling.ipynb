```python

patients = ["p1", "p2", "p3", "p4"]
arrival_times = [0, 4, 5, 6]
burst_times = [24, 3, 3, 12]
priorities = [3, 1, 4, 2]

num_patients = len(patients)

# Initialize arrays for waiting times and turnaround times
waiting_times_fcfs = [0] * num_patients
turnaround_times_fcfs = [0] * num_patients

waiting_times_sjf = [0] * num_patients
turnaround_times_sjf = [0] * num_patients

waiting_times_ps = [0] * num_patients
turnaround_times_ps = [0] * num_patients

waiting_times_rr = [0] * num_patients
turnaround_times_rr = [0] * num_patients

# FCFS 
for i in range(1, num_patients):
    waiting_times_fcfs[i] = turnaround_times_fcfs[i-1] + burst_times[i-1]
    if arrival_times[i] > turnaround_times_fcfs[i]:
        waiting_times_fcfs[i] = 0
    else:
        waiting_times_fcfs[i] -= arrival_times[i]
    turnaround_times_fcfs[i] = waiting_times_fcfs[i] + burst_times[i]

# SJF 
sorted_patients_sjf = sorted(range(num_patients), key=lambda k: burst_times[k])
for i in range(1, num_patients):
    patient_index = sorted_patients_sjf[i]
    waiting_times_sjf[patient_index] = max(0, turnaround_times_sjf[sorted_patients_sjf[i-1]] - arrival_times[patient_index])
    turnaround_times_sjf[patient_index] = waiting_times_sjf[patient_index] + burst_times[patient_index]

# PS 
sorted_patients_ps = sorted(range(num_patients), key=lambda k: priorities[k])
for i in range(1, num_patients):
    patient_index = sorted_patients_ps[i]
    waiting_times_ps[patient_index] = max(0, turnaround_times_ps[sorted_patients_ps[i-1]] - arrival_times[patient_index])
    turnaround_times_ps[patient_index] = waiting_times_ps[patient_index] + burst_times[patient_index]

# RR
time_quantum = 4
remaining_burst_time_rr = burst_times.copy()
current_time_rr = 0
queue_rr = []

while True:
    for i in range(num_patients):
        if arrival_times[i] <= current_time_rr and remaining_burst_time_rr[i] > 0:
            if remaining_burst_time_rr[i] <= time_quantum:
                current_time_rr += remaining_burst_time_rr[i]
                turnaround_times_rr[i] = current_time_rr
                remaining_burst_time_rr[i] = 0
            else:
                current_time_rr += time_quantum
                remaining_burst_time_rr[i] -= time_quantum
            while queue_rr and arrival_times[queue_rr[0]] <= current_time_rr:
                queue_rr.pop(0)
        elif arrival_times[i] > current_time_rr:
            current_time_rr += 1
        if sum(remaining_burst_time_rr) == 0:
            break
    if sum(remaining_burst_time_rr) == 0:
        break

for i in range(num_patients):
    waiting_times_rr[i] = turnaround_times_rr[i] - arrival_times[i] - burst_times[i]

# Calculate average waiting time and average turnaround time
avg_waiting_fcfs = sum(waiting_times_fcfs) / num_patients
avg_turnaround_fcfs = sum(turnaround_times_fcfs) / num_patients

avg_waiting_sjf = sum(waiting_times_sjf) / num_patients
avg_turnaround_sjf = sum(turnaround_times_sjf) / num_patients

avg_waiting_ps = sum(waiting_times_ps) / num_patients
avg_turnaround_ps = sum(turnaround_times_ps) / num_patients

avg_waiting_rr = sum(waiting_times_rr) / num_patients
avg_turnaround_rr = sum(turnaround_times_rr) / num_patients

# Print results
print("FCFS Average Waiting Time:", avg_waiting_fcfs)
print("\nFCFS Average Turnaround Time:", avg_turnaround_fcfs)

print("\nSJF Average Waiting Time:", avg_waiting_sjf)
print("\nSJF Average Turnaround Time:", avg_turnaround_sjf)
print()
print("Priority Scheduling (PS) Average Waiting Time:", avg_waiting_ps)
print("\nPriority Scheduling (PS) Average Turnaround Time:", avg_turnaround_ps)
print()
print("Round Robin (RR) Average Waiting Time:", avg_waiting_rr)
print("\nRound Robin (RR) Average Turnaround Time:", avg_turnaround_rr)

# Determine the most suitable algorithm based on average waiting time and average turnaround time
suitable_algorithm = min([
    ("FCFS", avg_waiting_fcfs, avg_turnaround_fcfs),
    ("SJF", avg_waiting_sjf, avg_turnaround_sjf),
    ("PS", avg_waiting_ps, avg_turnaround_ps),
    ("RR", avg_waiting_rr, avg_turnaround_rr)
], key=lambda x: (x[1], x[2]))

print("\nMost Suitable Algorithm:", suitable_algorithm[0])
```
